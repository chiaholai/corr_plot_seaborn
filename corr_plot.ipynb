```python
import os
import numpy as np
import pandas as pd
np.random.seed(0)
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
import warnings
warnings.filterwarnings('ignore')

fd_dir = "C:/Users/Marco/Desktop/RefERP_plot"
os.chdir(fd_dir)

datname = 'MeanAmp_YA34&OA34.csv'
dat_dir = fd_dir + '/data/' + datname
dat = pd.read_csv(dat_dir)
dat

plotYA = sns.jointplot('RefAmb_YA','RefFail_YA',data=dat,kind='reg',color="tab:blue",size=7,xlim=(-3,3),ylim=(-6,6),scatter_kws={"color":"tab:blue","alpha":0.7,"s":70},marginal_kws=dict(bins=np.linspace(-6,6,25))).set_axis_labels("Referential Ambiguity", "Referential Failure",fontsize=18)
plt.suptitle('Younger Adults',y=1.1, fontsize = 24)
plt.savefig('YAcorrplot.png', transparent=True)
plotOA = sns.jointplot('RefAmb_OA','RefFail_OA',data=dat,kind='reg',color="tab:blue",size=7,xlim=(-3,3),ylim=(-6,6),scatter_kws={"color":"tab:blue","alpha":0.7,"s":70},marginal_kws=dict(bins=np.linspace(-6,6,25))).set_axis_labels("Referential Ambiguity", "Referential Failure",fontsize=18)
plt.suptitle('Older Adults',y=1.1, fontsize = 24)
plt.savefig('OAcorrplot.png', transparent=True)
plt.show()
```
